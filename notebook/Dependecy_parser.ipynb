```python
# https://spacy.io/api/top-level#options
from spacy import displacy
import spacy
import settings
from pathlib import Path
```

```python
captions = ['A man preparing desserts in a kitchen covered in frosting.',
 'A chef is preparing and decorating many small pastries.',
 'A baker prepares various types of baked goods.',
 'a close up of a person grabbing a pastry in a container',
 'Close up of a hand touching various pastries.']
```

```python
nlp = spacy.load(settings.SPACY_MODEL)
```

```python
docs = [nlp(c) for c in captions]
```

```python
for doc in docs:
    displacy.render(doc, style='dep', jupyter=True, options={'compact': False})
```

```python
displacy.render(docs[2], style='dep', jupyter=True, options={'compact': False})
```

```python
svg = displacy.render(docs[2], style="dep", jupyter=False, options={'compact': False, 'distance': 90})
file_name = '-'.join([w.text for w in docs[2] if not w.is_punct]) + ".svg"
output_path = Path("images/" + file_name)
output_path.open("w", encoding="utf-8").write(svg)
```
